```python
#Import all required packages

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#Load CSV files

df = pd.read_csv(r'C:\Users\marya\Downloads\kc_house_data.csv')
```

```python
#Display the first 5 rows of the dataframe

df.head()
```

```python
#Check the missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} -{}%'.format(col, pct_missing))
```

```python
#Data Types for columns

df.dtypes
```

```python
df['price'] = df['price'].astype('int64')
```

```python
df.head()
```

```python
#Order Data

df = df.sort_values(by=['price'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
plt.scatter(x=df['price'], y=df['sqft_living'])
plt.title('price vs sqft_living')
plt.xlabel('sqft_living')
plt.ylabel('price')
plt.show()
```

```python
df.head()
```

```python
sns.regplot(x='price',y='sqft_living',data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
#Correlation Matrix between all numeric columns

df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix ,annot=True)
plt.title('Correlation')
plt.xlabel('House Features')
plt.ylabel('House Features')
plt.show()
```
